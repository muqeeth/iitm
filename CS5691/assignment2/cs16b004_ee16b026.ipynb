```python
# CodeWrite 
#Write logistic regression code from scratch. Use gradient descent.
# Only write functions here

def kernel_matrix(X,Y,kernel,kernel_param = 1):
    if(kernel == 'linear'):
        return np.matmul(X,Y.T)
    elif (kernel=='poly'):
        return (1+np.matmul(X,Y.T))**(kernel_param)
    elif (kernel=='rbf'):
        l = np.sum(np.abs(X)**2,axis=1)
        r = np.sum(np.abs(Y)**2,axis=1)
        p = -(2*np.matmul(X,Y.T))
        return np.exp(-kernel_param*(np.add(np.add(p,r).T,l).T))
    
def grad_descent(K,Y_train,reg_param):
    eta_range = [0.01]
    R_alpha = 10000
    alpha = np.zeros(len(Y_train))
    best_eta = 1
    for eta in eta_range:
        temp_alpha = np.zeros(len(Y_train))
        for i in range(100):
            l = (np.exp(-np.multiply(np.matmul(temp_alpha,K),Y_train))/
                       (1+np.exp(-np.multiply(np.matmul(temp_alpha,K),Y_train))))
            R_grad = (np.sum(np.multiply(l,-np.multiply(Y_train,K))/len(Y_train),axis=-1)
                      +reg_param*np.matmul(K,temp_alpha))
#             R_grad = np.zeros(len(Y_train))
#             for ind in range(len(Y_train)) :
#                 l = (np.exp(-np.matmul(temp_alpha,K.T[ind])*Y_train[ind])/
#                            (1+np.exp(-np.matmul(temp_alpha,K.T[ind]*Y_train[ind]))))
#                 R_grad +=  -l*Y_train[ind]*K.T[ind]
#             R_grad = R_grad/len(Y_train)
#             R_grad = R_grad + reg_param*np.matmul(K,temp_alpha)
            temp_alpha =  temp_alpha - eta*(R_grad)
            alpha = temp_alpha
        temp_R = (np.sum(np.log(np.exp(-np.multiply(np.matmul(temp_alpha,K),Y_train)) + 1)/len(Y_train),axis=-1) 
                  +reg_param*(np.matmul(np.matmul(temp_alpha,K),temp_alpha))/2)
#             temp_R =  0
#             for ind in range(len(Y_train)) :
#                 l = np.log(1+np.exp(-np.matmul(temp_alpha,K.T[ind]*Y_train[ind])))
#                 temp_R +=  l
#             temp_R  = temp_R/len(Y_train)
#             temp_R  += reg_param*(np.matmul(np.matmul(temp_alpha,K),temp_alpha))/2
        if(temp_R<=R_alpha):
            R_alpha = temp_R
            alpha = temp_alpha
            best_eta = eta

#     print('best eta is {}'.format(best_eta))
    return alpha
    
def train_pred_logistic_regression(X_train, Y_train, X_test, kernel='linear', reg_param=0., 
                                   kernel_param=1., num_iter_gd=100):
    K = kernel_matrix(X_train,X_train,kernel,kernel_param)
    alpha = grad_descent(K,Y_train,reg_param)
    l = kernel_matrix(X_train,X_test,kernel,kernel_param)
    Y_test_pred = np.sum(np.multiply(alpha[:,np.newaxis],l),axis = 0)
    Y_test_pred = 1/(1+np.exp(-Y_test_pred))
    Y_test_pred[Y_test_pred>=0.5] = 1
    Y_test_pred[Y_test_pred<0.5] = -1
    return Y_test_pred
    
def return_best_hyperparam(kernel,X_train,Y_train,X_valid,Y_valid): 
    # give appropriate arguments, return appropriate variables
    error = 1
    best_reg_param = 1
    best_kernel_param = 1
    reg_param_range = [1e-8,1e-6,1e-4,1e-2,1]
    kernel_param_range = [1e-2,1e-1,1,1e1]
#     reg_param_range = [1e-2]
#     kernel_param_range = [ 1e1]
    if(kernel=='linear'):
        for reg_param in reg_param_range:
            Y_valid_pred = train_pred_logistic_regression(X_train, Y_train, X_valid,'linear', 
                                                          reg_param,1,100)
            temp_error = len(np.where(Y_valid !=Y_valid_pred)[0])/len(Y_valid)
#             print('error for linear kernal reg_param {} is {}'.format(reg_param,temp_error))
            if(temp_error<=error):
                error = temp_error
                best_reg_param = reg_param
    else:
        for kernel_param in kernel_param_range:
            for reg_param in reg_param_range:    
                Y_valid_pred = train_pred_logistic_regression(X_train, Y_train, X_valid,kernel, 
                                                              reg_param,kernel_param,100)
                temp_error = len(np.where(Y_valid !=Y_valid_pred)[0])/len(Y_valid)
#                 print('error for rbf kernel_param {} reg_param {} is {}'
#                       .format(kernel_param, reg_param,temp_error))
                if(temp_error<error):
                    error = temp_error
                    best_reg_param = reg_param
                    best_kernel_param = kernel_param
                    
    return best_kernel_param,best_reg_param
```

```python
# CodeWrite cell
# Write SVM classifier using SKlearn, write code for choosing best hyper parameters.
# write only functions here

def train_pred_svm(X_train,Y_train,X_test,reg_param,kernel='linear',kernel_param =1.0):
    if(kernel=='linear'):
        classif_algo =  SVC(C=reg_param,kernel='linear')
        classifier = classif_algo.fit(X_train,Y_train)
        Y_pred = classifier.predict(X_test)
        return Y_pred
    elif(kernel=='rbf'):
        classif_algo =  SVC(C=reg_param,kernel='rbf',gamma = kernel_param)
        classifier = classif_algo.fit(X_train,Y_train)
        Y_pred = classifier.predict(X_test)
        return Y_pred
    elif(kernel == 'poly'):
        classif_algo =  SVC(C=reg_param,kernel='poly',degree = int(kernel_param),coef0 =1 
                            ,gamma =1,max_iter = 10000)
        classifier = classif_algo.fit(X_train,Y_train)
        Y_pred = classifier.predict(X_test)
        return Y_pred
    
# def best_hyperparam(X_train,Y_train,X_valid,Y_valid,kernel='linear'):
#     reg_param_range = [1e-6,1e-4,1e-2,1,1e1]
#     kernel_param_range = [1e-3,1e-2,1e-1,1,1e1]
#     deg_param_r
#     best_reg_param = 0
#     best_kernel_param = 1
#     best_error = 1
#     for reg_param in reg_param_range:
#         for kernel_param in kernel_param_range:
#             Y_valid_pred = np.zeros(len(Y_valid))
#             if(kernel=='linear'):
#                 Y_valid_pred = train_pred_svm(X_train,Y_train,X_valid,reg_param)
#             else:
#                 Y_valid_pred = train_pred_svm(X_train,Y_train,X_valid,reg_param,kernel,kernel_param)
#             error = len(np.where(Y_valid_pred!=Y_valid)[0])/len(Y_valid)
#             if(error<best_error):
#                 best_error = error
#                 best_reg_param = reg_param
#                 best_kernel_param = kernel_param
#     return best_reg_param,best_kernel_param
def best_hyperparam(X_train,Y_train,X_valid,Y_valid,kernel='linear'):
    reg_param_range = [1e-6,1e-4,1e-2,1,1e1]
    kernel_param_range = [1e-3,1e-2,1e-1,1,1e1]
    degree_param_range = [2,3,4,5]
    best_reg_param = 0
    best_kernel_param = 100
    best_error = 1
    if(kernel == 'linear'):
        for reg_param in reg_param_range:
            Y_valid_pred = np.zeros(len(Y_valid))
            Y_valid_pred = train_pred_svm(X_train,Y_train,X_valid,reg_param)
            error = len(np.where(Y_valid_pred!=Y_valid)[0])/len(Y_valid)
            if(error<best_error):
                best_error = error
                best_reg_param = reg_param
                best_kernel_param = 1
    elif(kernel == 'rbf'):
        for reg_param in reg_param_range:
            for kernel_param in kernel_param_range:
                Y_valid_pred = np.zeros(len(Y_valid))
                Y_valid_pred = train_pred_svm(X_train,Y_train,X_valid,reg_param,kernel,kernel_param)
                error = len(np.where(Y_valid_pred!=Y_valid)[0])/len(Y_valid)
                if(error<best_error):
                    best_error = error
                    best_reg_param = reg_param
                    best_kernel_param = kernel_param
    else:
        for reg_param in reg_param_range:
            for degree in degree_param_range:
                Y_valid_pred = np.zeros(len(Y_valid))
                Y_valid_pred = train_pred_svm(X_train,Y_train,X_valid,reg_param,kernel,degree)
                error = len(np.where(Y_valid_pred!=Y_valid)[0])/len(Y_valid)
                if(error<best_error):
                    best_error = error
                    best_reg_param = reg_param
                    best_kernel_param = degree
    return best_reg_param,best_kernel_param
            
    
```

```python
# CodeWrite cell
# Write Random Forest classifier assuming access to a decision tree learner, 
# write code for choosing best node size to stop splitting.
# write only functions here
def decision_tree(X_train,Y_train,X_valid):
    clf = DecisionTreeClassifier()
    clf.fit(X_train,Y_train)
    Y_valid = clf.predict(X_valid)
    return Y_valid,clf
def best_param(X_train,Y_train,X_valid,Y_valid):
    best_features_list = []
    best_classifier = []
    best_error = 1
    best_number = 0
    number_trees = np.arange(10,1000,100) + 1
    for number in number_trees:
        features_list = []
        classifier= []
        Y_valid_pred = np.zeros(len(Y_valid))
        for i in range(number):
            points = np.arange(0,np.shape(X_train)[0],1)
            features = np.arange(0,np.shape(X_train)[1],1)
            np.random.shuffle(features)
            np.random.shuffle(points)
            features = features[0:int(np.shape(features)[0]/2)]
            features_list.append(features)
            points = points[0:int(np.shape(points)[0]/2)]
            X = X_train[points]
            Y = Y_train[points]
            X = X.T[features].T
            X_v = X_valid.T[features].T
            y,cl = decision_tree(X,Y,X_v)
            classifier.append(cl)
            Y_valid_pred +=y
        Y_valid_pred[Y_valid_pred>=0] = 1
        Y_valid_pred[Y_valid_pred<0] = -1
        error = len(np.where(Y_valid_pred!=Y_valid)[0])/len(Y_valid)
        if(error<=best_error):
            best_error = error
            best_features_list = features_list.copy()
            best_classfier = classifier.copy()
            best_number = number
    return best_features_list,best_classfier,best_number
def prediction(features,classifier,number,X_test):
    Y_pred = np.zeros(len(X_test))
    for i in range(number):
        X = X_test.T[features[i]].T
        Y_pred+=classifier[i].predict(X)
    Y_pred[Y_pred>=0] = 1
    Y_pred[Y_pred<0] = -1
    return Y_pred
```
